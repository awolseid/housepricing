```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Dataset

```python
housing = pd.read_csv("housing.csv")
housing
```

```python
class LikeBuiltinData:
    def __init__(self, dataframe: pd.DataFrame = housing):
        self.dataframe = dataframe
        
    @property
    def data(self):
        return housing.iloc[: , :-1].to_numpy()
    
    @property
    def target(self):
        return housing["Price"].to_numpy()
    
    @property
    def feature_names(self):
        return housing.iloc[: , :-1].columns.tolist()
    
    @property
    def DESCR(self):
        return "This is the Boston Housing dataset."

    
price_dataset = LikeBuiltinData()
price_dataset.feature_names
```

```python
print(price_dataset.data)
```

```python
# print(price_dataset.target)
```

## Prepare Dataset

```python
dataset = pd.DataFrame(price_dataset.data, columns = price_dataset.feature_names)
dataset
```

```python
dataset["Price"] = price_dataset.target
dataset
```

```python
### Independent and Dependent Features

X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
### Standardize dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
import pickle

pickle.dump(scaler, open("scaling.pkl", "wb"))
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) # in the test dataset, no fit_transform
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression
```

```python
regression.fit(X_train, y_train)
```

```python
regression.get_params()
```

```python
### print coefficients and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# prediction with Test Data
reg_pred = regression.predict(X_test)
reg_pred
```

```python
### plot scatter plot for the prediction
plt.scatter(reg_pred, y_test);
```

```python
residuals = y_test - reg_pred
residuals
```

```python
### plot of residuals
sns.displot(residuals, kind = "kde");
```

```python
### scatter plot of predictions and residuals
plt.scatter(reg_pred, residuals)
```

```python
### Performance metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
### R Square and adjusted R Square
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, reg_pred)
print(r2_score)
```

```python
adj_r2 = 1 - (1 - r2_score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
print(adj_r2)
```
